```python
import pandas as pd

# Load datasets
df1 = pd.read_excel(r"C:\Users\hamza\Downloads\Mushroom growth tracking (1).xlsx")
df2 = pd.read_excel(r"C:\Users\hamza\Downloads\final cycle 1.xlsx")

# Convert date columns to datetime format
df1['date'] = pd.to_datetime(df1['date'])
df2['time'] = pd.to_datetime(df2['time'])

# Dictionary to store the last timestamp for each id in the first dataset
last_timestamps = {}

# Iterate through rows in the first dataset
for index, row in df1.iterrows():
    # Extract relevant information
    id_value = row['id']
    timestamp = row['date']

    # Add a condition to consider only rows before the 3rd month
    if timestamp.month > 3:
        # Check if the id has appeared before
        if id_value in last_timestamps:
            # Check if the previous row has the same id
            if df1.at[index - 1, 'id'] == id_value:
                # Filter rows in the second dataset based on the time window
                mask = (df2['time'] >= last_timestamps[id_value]) & (df2['time'] <= timestamp)
                relevant_rows = df2[mask]
            else:
                # If the previous row has a different id, filter rows based on minutes
                mask = (df2['time'].dt.minute == timestamp.minute)
                relevant_rows = df2[mask]

            # Calculate average of field1 and field2
            if not relevant_rows.empty:


                # Add new columns 'hum', 'dht', 'ds' based on the second dataset
                df1.at[index, 'hum'] = relevant_rows['hum'].mean()
                df1.at[index, 'dht'] = relevant_rows['dht'].mean()
                df1.at[index, 'ds'] = relevant_rows['ds'].mean()

        else:
            # If id_value is not in last_timestamps, fill 'hum', 'dht', 'ds' based on minutes
            mask = (df2['time'].dt.minute == timestamp.minute)
            relevant_rows = df2[mask]

            if not relevant_rows.empty:
                df1.at[index, 'hum'] = relevant_rows['hum'].mean()
                df1.at[index, 'dht'] = relevant_rows['dht'].mean()
                df1.at[index, 'ds'] = relevant_rows['ds'].mean()

        # Update the last timestamp for the current id
        last_timestamps[id_value] = timestamp

# Save the modified dataset to a new Excel file
df1.to_excel("output_excel03_with_columns_updated.xlsx", index=False)
```
